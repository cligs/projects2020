```python
wdir = "/home/jose/Dropbox/Doktorarbeit/publications/202006_DH/data/theatreclassique/"
```

```python
corpus_df = load_data.load_corpus(wdir, wsdir="txt/", sep="\t", verbose=True, min_MFF=0, max_MFF=False, lowercase=True,
                keep_punctuation=True)
```

```python
corpus_rel_df = corpus_df.loc[:].div(corpus_df.sum(axis='columns'), axis="index")
```

```python
corpus_rel_df.head()
```

```python
from sklearn.decomposition import LatentDirichletAllocation
```

```python
corpus_rel_zscores_df = text2features.calculate_zscore(corpus_rel_df)
```

```python
corpus_rel_zscores_df = corpus_rel_zscores_df + (corpus_rel_zscores_df.values.min()*-1)
```

```python
corpus_rel_tfidf_df = text2features.calculate_tfidf_sklearn(corpus_rel_df)
```

```python
corpus_rel_zscores_df.head()
```

```python
corpus_rel_tfidf_df.head()
```

```python
lda = LatentDirichletAllocation(n_topics = 300, max_iter = 10, learning_method = 'online')
lda.fit(corpus_rel_zscores_df.iloc[:,500:5000])
```

lda_model = LatentDirichletAllocation(n_topics=20,               # Number of topics
                                      max_iter=10,               # Max learning iterations
                                      learning_method='online',   
                                      random_state=100,          # Random state
                                      batch_size=128,            # n docs in each learning iter
                                      evaluate_every = -1,       # compute perplexity every n iters, default: Don't
                                      n_jobs = -1,               # Use all available CPUs
                                     )
lda_output = lda_model.fit_transform(corpus_rel_df.iloc[:,:3000])

```python
def display_topics(model, feature_names, no_top_words):
    topics_names_lt = []
    for topic_idx, topic in enumerate(model.components_):
        topics_names_lt.append("Topic " + str(topic_idx) + ": \n" + " ".join([feature_names[i] for i in topic.argsort()[:-no_top_words - 1:-1]]))
    return topics_names_lt

no_top_words = 3
topics_names_lt = display_topics(lda, corpus_rel_zscores_df.iloc[:,500:5000].columns.tolist(), no_top_words)
```

```python
#lda.transform(corpus_rel_zscores_df.iloc[:,500:5000])
```

```python
topics_texts_df = pd.DataFrame(lda.transform(corpus_rel_zscores_df.iloc[:,500:5000]), index = corpus_rel_df.index.tolist(), columns=topics_names_lt)
```
